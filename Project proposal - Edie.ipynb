**Analysis 1**

**Research question: Is there correlation between self-reported happiness and whether the social needs with friends have been met ?**

**Variables:**

These data are directly related to the research question. All are ordinal data from 2021 cross-sectional survey. 
WELLNESS_subjective_happiness_scale_always_happy: 8 categorical section responding Some people are generally very happy. They enjoy life regardless of what is going on, getting the most out of everything. To what extent does this characterization describe you?

CONNECTION_preference_time_friends_grouped:  5 categorical section responding to How much time per week would you like to spend socializing with others from the following groups? - Friends

CONNECTION_social_time_friends_p7d_grouped: 5 categorical section responding to ‘In the PAST WEEK, how many hours in total did you spend socializing with others from the following groups? - Friends’ 

**Visualization:**

Use bar plots. Bar plots are efficient to visualize categorical data by counting frequency of each category and represent the frequency by the height of each bar. This can give us an intuitive assessment. 

**Analysis plan:**

Clean the data that selected NaN in either preference time and social time as they cannot make comparisons. Then create a new variable Social_time_met using CONNECTION_preference_time_friends_grouped and CONNECTION_social_time_friends_p7d_grouped, as both of them have the same categories, we can rank their choices and calculate; if actual time ≥ preferred time, code it as “Met” (1).
If actual time < preferred time, code it as “Not Met” (0). Set the alpha value to 0.05. Use python to construct out spearman correlation, and calculate out the p-value. If the p value is smaller than the alpha value, then the R value obtained is statistically significant. if the R is close to +1, then it indicates a strong positive correlation and close to -1 indicates a strong negative correlation. 

**Hypothesis:**

Null hypothesis: there’s no correlation between the self-reported happiness and whether the social needs with friends have been met.

Alternative hypothesis: there is a correlation between the self-reported happiness and whether the social needs with friends have been met.

**Possible outcome:** 

Depending on whether the null hypothesis will be rejected by the result, we can determine if there is a significant correlation between self-reported happiness and whether the social needs with friends have been met.
